# EDA - Spooky Author Identification

Running some exploratory analysis on the training set to see what we've got to work with.

```python
import pandas as pd
```

```python
train_df = pd.read_csv("train.csv")
train_df.head()
```

## How many sentences were written by each author?

```python
train_df.groupby(["author"]).size().reset_index(name='counts')
```

## What's the average length of a sentence written by each author?

```python
train_df["char_count"] = train_df["text"].str.len()
train_df.groupby(["author"]).agg({'char_count': ['sum', 'mean']})
```

```python
import nltk
from collections import Counter

def count_pos(pos_to_count, text):
    c = Counter()
    for term,pos in nltk.pos_tag(nltk.word_tokenize(text)):
        c[pos] +=1    
    return c[pos_to_count]
    
head = train_df.copy().head()
for pos in ["NN", "VB"]:
    head["{0}_count".format(pos)] = head["text"].apply(lambda text: count_pos(pos, text))
    train_df["{0}_count".format(pos)] = train_df["text"].apply(lambda text: count_pos(pos, text))
5555    

# nltk.pos_tag(nltk.word_tokenize(head.text[0]))
```
